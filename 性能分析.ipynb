```python
import time
#导入工具包
from tools.preprocess import preprocess, distance, delete_file
from tools.data_loader import data_loader
from tools.results_process import results_process
from tools.predict import predict
#使用sklearn
from sklearn.cluster import KMeans, DBSCAN, OPTICS, AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
#数据预处理
df = preprocess("lng2.csv", "lng2_stop_spots.csv")
print(df)
```

```python
#数据加载
loader = data_loader("lng2_stop_spots.csv")
data, coord = loader.load()
```

```python
plt.figure(figsize=(9,7),dpi=100)
times = []
for i in range(8):
    i = i + 1
    start =time.time()
    DBSCAN(eps=0.05, min_samples=3, n_jobs=i).fit(coord)
    end = time.time()
    times.append(end-start)
plt.plot([1, 2, 3, 4, 5, 6, 7, 8],times,color = 'r',label = 'DBSCAN')
plt.show()
```

```python
plt.figure(figsize=(9,7),dpi=100)
times = []
for i in range(8):
    i = i + 1
    start =time.time()
    OPTICS(min_samples=10, n_jobs=i).fit(data)
    end = time.time()
    times.append(end-start)
plt.plot([1, 2, 3, 4, 5, 6, 7, 8],times,color = 'r',label = 'Optics')
plt.show()
```
